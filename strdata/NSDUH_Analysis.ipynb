```python
import numpy as np

for col1 in use_columns:
    for col2 in diag_columns:
        df[f'{col1}_{col2}'] = df.apply(
            lambda row: np.nan if pd.isna(row[col1]) or pd.isna(row[col2]) else int(row[col1] == 1 and row[col2] == 1),
            axis=1
        )
```
