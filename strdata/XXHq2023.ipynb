```python
import pandas as pd
import numpy as np
import seaborn as sns

# Set pandas to display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('XXHq 2023 Cleaned.csv')
df
```

```python
df.info()
```

```python
df_quality = df.iloc[:, 6:21]
df_quality
```

```python
quality_cols = list(df_quality.columns)
quality_cols
```

```python
df_use = df.iloc[:, 22:45]
df_use
```

```python
use_cols = list(df_use.columns)
use_cols
```

```python
df_mental_health = df.iloc[:,50:63]
df_mental_health
```

```python
mental_cols = list(df_mental_health.columns)
mental_cols
```

```python
# Filter out columns in your dataframe
df_use_qual = df[use_cols + quality_cols]

# Compute the correlation matrix
correlation_matrix = df_use_qual.corr()

# Extract correlations between use_columns and comb_columns
corr_use_qual = correlation_matrix.loc[use_cols, quality_cols]

# Display the correlation results
corr_use_qual
```

```python
# Plot correlation heatmap
sns.heatmap(corr_use_qual, 
            cmap="YlGnBu", 
            annot=False,)

# Display heatmap
```
